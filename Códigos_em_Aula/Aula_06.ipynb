```python
def soma_numero(n, lista):
    soma = 0
    for intera in range(n):
        soma += lista[intera]
    return soma

def main():
    lista = [1,2,3,4,5,6,7,8,9]
    print(soma_numero(3, lista))

if __name__ == "__main__":
    main()
```

```python
def fatorial(n):
    # 3! = 3x2x1
    if n == 1:
        return 1
    return n * fatorial(n-1)

print('resultado: ', fatorial(6))
```

```python
def fibonacci(n):
    # 3! = 3x2x1
    if n == 1 or n == 0:
        return 1
    elif n < 0:
        return 0
    return fibonacci(n-1) + fibonacci(n-2)

print('resultado: ', fibonacci(-1))
```
