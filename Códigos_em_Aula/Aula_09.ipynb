```python
notas = pd.Series([7.5, 8.0, 9.2, 6.8, 10.0], index=['Diego', 'Marcos', 'Isac', 'Cauã', 'Carlúcio'], name='Nota Final')
print(notas)
```

```python
# localização do dado
print(notas.iloc[0])

# print com relação ao index
print(notas['Carlúcio'])
```

```python
aprovados = notas[notas >= 7]
print(aprovados)
```

```python
# ── Estatísticas ──
print(notas.mean())
print(notas.median())
print(notas.std())
print(notas.min())
print(notas.max())
print(notas.sum())
print()
#print(notas.describe())
```

```python
print(f'Notas normalizadas:\n{notas *10}')
```

```python
def classifica(n):
    if n >= 9:
        return 'Aprovado com nota excelente'
    if n >= 7:
        return 'Aprovado'
    return 'Reprovado'

# O natural seria:
#       classifica(notas)

situacao = notas.apply(classifica)  #FAZ ISSO EMBUTIDO, FAZ PARTE DA RECURSÃO, MUITO MAIS VELOZ
print(situacao)
```
