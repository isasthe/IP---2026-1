```python
# Exercício 4
#  Desenvolva um programa que leia o valor da hora trabalhada e a quantidade de horas no mês, calcule o salário bruto e depois aplique os descontos
hora_sal = int(input("Qual o valor da hora trabalhada? "))
hora_tra = int(input("Qual a quantidade de horas do salário proposto? "))
bolsa = int(input("Qual o valor da bolsa atual? "))

## Cálculos
sal_bru = hora_sal * hora_tra
if (sal_bru <= 1900):
  ir = 0
  inss = sal_bru * 0.1
  sal_liq = sal_bru - inss - ir
elif (sal_bru <= 2800):
  ir = sal_bru * 0.075
  inss = sal_bru * 0.1
  sal_liq = sal_bru - inss - ir
elif (sal_bru <= 3700):
  ir = sal_bru * 0.15
  inss = sal_bru * 0.1
  sal_liq = sal_bru - inss - ir
else:
  ir = sal_bru * 0.2
  inss = sal_bru * 0.1
  sal_liq = sal_bru - inss - ir

## Comparação com bolsa
if (sal_liq < bolsa):
  comp4 = ("menor que a bolsa")
  dif_bolsa = bolsa - sal_liq
elif (sal_liq > bolsa):
  comp4 = ("maior que a bolsa")
  dif_bolsa = sal_liq - bolsa
else:
  comp4 = ("igual à bolsa")

## Resultados
print("")
print(f"O salário bruto é: R${sal_bru:.2f}")
print(f"O IR é: R${ir:.2f}")
print(f"O INSS é: R${inss:.2f}")
print(f"O salário líquido é: R${sal_liq:.2f}")
if (dif_bolsa != 0):
  print(f"O salário líquido é R${dif_bolsa:.2f} {comp4}")
else:
  print(f"O salário líquido é {comp4}")
```

```python
# Exercício 5
saque = int(input("Qual o valor a ser sacado? "))

cem_tr = 100
cinq_tr = 50
dez_tr = 10
cin_tr = 5
um_tr = 1

ced100 = 0
ced50 = 0
ced10 = 0
ced5 = 0
ced1 = 0


while ((saque - cem_tr) > 0):
  saque -= cem_tr
  ced100 += 1
while ((saque - cinq_tr) > 0):
  saque -= cinq_tr
  ced50 += 1
while ((saque - dez_tr) > 0):
  saque -= dez_tr
  ced10 += 1
while ((saque - cin_tr) > 0):
  saque -= cin_tr
  ced5 += 1
while (saque > 0):
    saque -= 1
    ced1 += 1

cedulas = ced100 + ced50 + ced10 + ced5 + ced1
print("")
print(f"A menor quantidade possível de cédulas usadas foi {cedulas}")
print("A quantidade de cada cédula foi: ")
if (ced100 > 0):
  print(f"  Cédulas de 100: {ced100}")
if (ced50 > 0):
  print(f"  Cédulas de 50: {ced50}")
if (ced10 > 0):
  print(f"  Cédulas de 10: {ced10}")
if (ced5 > 0):
  print(f"  Cédulas de 5: {ced5}")
if (ced1 > 0):
  print(f"  Cédulas de 1: {ced1}")
```

```python
# Exercício 7
num_x = int(input("Qual o primeiro número? "))
num_y = int(input("Qual o segundo número? "))
op = input("Qual a operação desejada? (soma - subtração - multiplicação - divisão) ")
print("")

if (op == ("soma")):
  print(f"{num_x} + {num_y} = {num_x + num_y}")
elif (op == ("subtração")):
  print(f"{num_x} - {num_y} = {num_x - num_y}")
elif (op == ("multiplicação")):
  print(f"{num_x} x {num_y} = {num_x * num_y}")
elif (op == ("divisão")):
  print(f"{num_x} / {num_y} = {num_x / num_y}")
else:
  print("Erro. Digite um valor válido")
```

```python
# Exercício 8

lados = []

for m in range(3):
  lados.append(int(input(f"Qual o tamanho do segmento {m + 1}? ")))

soma_lados = sum(lados)
lados_set = set(lados)
if ((max(lados)) >= soma_lados - max(lados)):
  print("Não podem formar um triângulo")
elif (len(lados_set) == 3):
  print("É um triângulo escaleno")
elif (len(lados_set) == 2):
  print("É um triângulo isósceles")
else:
  print("É um triângulo equilátero")
```
