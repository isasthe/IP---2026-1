```python
# Exercício 1
n = int(input("Escolha um número inteiro positivo N: "))
cont_n = 0
sum_cont_n = 0
while (cont_n < n):
  cont_n += 1
  print(cont_n)
  sum_cont_n += cont_n
print("")
print(f"A soma total é: {sum_cont_n}")
```

```python
# Exercício 2
notas = []
cont_notas = 0
while (True):
  cont_notas += 1
  nota = int(input(f"Qual a nota {cont_notas}? "))
  if (nota == -1):
    break
  elif (-1 <= nota <= 10):
    notas.append(nota)

print("")
print(f"Quantidade de notas válidas: {len(notas)}")
media = sum(notas) / len(notas)
print(f"Média das notas: {media:.2f}")
print(f"A maior nota é: {max(notas)}")
print(f"A menor nota é: {min(notas)}")
print("")

## verificar quantos estão acima da média
cont_notas = 0
for n in range(len(notas)):
  if (notas[n] >= media):
    cont_notas += 1
print(f"Quantidade de notas acima da média: {cont_notas}")
```

```python
# Exercício 4

def tabuada(num_n, alcance):
  for multiplicador in range (1, alcance+1):
    print(f"{num_n} x {multiplicador} = {num_n * multiplicador}")

def main():
  num_n = int(input("Qual o número inteiro a ser multiplicado? "))
  alcance = int(input("Escolha até onde irá a tabuada (ex: 10): "))
  print("")
  tabuada(num_n, alcance)

if __name__ == "__main__":
    main()
```

```python
# Exercício 5
def temperatura():
  temp = []
  for n in range(7):
    temp.append(float(input(f"Qual a temperatura do {n + 1}º dia? ")))
  return temp

def calc_temp(temp):
  media = sum(temp) / len(temp)
  max_temp = max(temp)
  min_temp = min(temp)
  return media, max_temp, min_temp

def main():
  temp = temperatura()
  media, max_temp, min_temp = calc_temp(temp)
  print("")
  print(f"A temperatura média foi: {media:.1f}°C")
  print(f"A maior temperatura foi: {max_temp:.1f}°C")
  print(f"A menor temperatura foi: {min_temp:.1f}°C")

if __name__ == "__main__":
  main()
```

```python
# Exercício 6

n_num = int(input("Escolha um número N: "))
primos = []
cont = 0

for contador in range(2, n_num+1):
  for div in range(2, contador):
    if (contador % div == 0):
      cont += 1
    if cont == 1:
      break
  if cont == 0:
    primos.append(contador)
  cont = 0

print("")
print(primos)
print(f"Ao todo, de 2 a {n_num} a quantidade de números primos é: {len(primos)}")
print(f"Ao todo, a soma dos números primos de 2 a {n_num} é: {sum(primos)}")
```

```python
# Exercício 7
import random
tentativas = 0
num_aleatorio = random.randrange(2, 100)
while (True):
  palpite = int(input("Diga um palpite sobre qual número aleatório entre 1 e 100 estou pensando: "))
  tentativas += 1
  if (palpite == num_aleatorio):
    print(f"Parabéns! Você acertou em {tentativas} tentativas. O número era: {num_aleatorio}")
    break
  elif (palpite < num_aleatorio):
    print(f"Muito baixo! (Seu palpite foi: {palpite})")
  elif (palpite > num_aleatorio):
    print(f"Muito alto! (Seu palpite foi: {palpite})")
```

```python
# Exercício 8

numero_equi = int(input("Quantos equipamentos serão cadastrados? "))
equipamentos = []
for _ in range(numero_equi):
  nome_equi = input("Qual o nome do equipamento? ")
  valor_equi = int(input(f"Qual o valor do(a) {nome_equi}? "))
  equipamentos.append((nome_equi, valor_equi))

print("")
for nome_equi, valor_equi in equipamentos:
  print(f"{nome_equi}, de valor: R${valor_equi:.2f}")
print("")
total_equi = sum(valor_equi for _, valor_equi in equipamentos)
print(f"O valor total do inventário é R${total_equi:.2f}")
print("")
max_equi = max(equipamentos, key=lambda x: x[1])
nome_equi, valor_equi = max_equi
print(f"O equipamento mais caro é o(a) {nome_equi}, de valor R${valor_equi:.2f}")
min_equi = min(equipamentos, key=lambda x: x[1])
nome_equi, valor_equi = min_equi
print(f"O equipamento mais barato é o(a) {nome_equi}, de valor R${valor_equi:.2f}")
```
